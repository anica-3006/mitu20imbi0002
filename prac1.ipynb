```python
import matplotlib.pyplot as plt
from abc import ABC, abstractmethod

class BioData(ABC):
    # What did you do?
    def __init__(self, name, data):
        self.name = name
        self.data = data

    @abstractmethod
    def analyze(self):
        pass

    @abstractmethod
    def visualize(self):
        pass

class ProteinSequence(BioData):
    def __init__(self, name, sequence):
        super().__init__(name, sequence)

    def analyze(self):
        print("\nAnalyze Protein Sequence...")
        print("Name: " + self.name)
        print("Length of the sequence: " + str(len(self.data)))
        print("Unique Amino acids: " + str(len(set(self.data))))

    def visualize(self):
        print("\nVisualize Protein Sequence...")
        amino_acid_count = {aa: self.data.count(aa) for aa in set(self.data)}
        plt.bar(amino_acid_count.keys(), amino_acid_count.values())
        plt.title("Amino Acid Distribution")
        plt.show()

class DnaSequence(BioData):
    def __init__(self, name, sequence):
        super().__init__(name, sequence)

    def analyze(self):
        print("\nAnalyze DNA Sequence...")
        print("Name: " + self.name)
        print("Length of the sequence: " + str(len(self.data)))
        print("Unique Nucleotides: " + str(len(set(self.data))))

    def visualize(self):
        print("\nVisualize DNA Sequence...")
        nucleotide_count = {nt: self.data.count(nt) for nt in set(self.data)}
        plt.bar(nucleotide_count.keys(), nucleotide_count.values())
        plt.title("Nucleotide Distribution")
        plt.show()

def choice():
    while True:
        print("\nBioData Analysis")
        print("1. Analyze Protein Sequence")
        print("2. Visualize Protein Sequence")
        print("3. Analyze DNA Sequence")
        print("4. Visualize DNA Sequence")
        print("5. Exit")

        choice = input("Enter your choice (1-5): ")

        if choice == '1':
            name = input("Enter protein name: ")
            sequence = input("Enter the protein sequence: ")
            protein = ProteinSequence(name, sequence)
            protein.analyze()
        elif choice == '2':
            name = input("Enter protein name: ")
            sequence = input("Enter the protein sequence: ")
            protein = ProteinSequence(name, sequence)
            protein.visualize()
        elif choice == '3':
            name = input("Enter gene name: ")
            sequence = input("Enter the DNA sequence: ")
            dna = DnaSequence(name, sequence)
            dna.analyze()
        elif choice == '4':
            name = input("Enter gene name: ")
            sequence = input("Enter the DNA sequence: ")
            dna = DnaSequence(name, sequence)
            dna.visualize()
        elif choice == '5':
            print("Exiting the program.")
            break
        else:
            print("Invalid choice, please try again.")

choice()
```
